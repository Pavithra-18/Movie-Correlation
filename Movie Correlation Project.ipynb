```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Reading the data
df = pd.read_csv('C:\\Users\\Pavithra\\Downloads\\archive (1)\\movies.csv')
```

```python
df.head()
```

```python
#Checking for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#Clearing the rows and columns with null values
cleaned_df = df.dropna()
```

```python
cleaned_df.info()
```

```python
print("Original DataFrame shape:", df.shape)
```

```python
print("Cleaned DataFrame shape:", cleaned_df.shape)
```

```python
#Checking for datatypes
cleaned_df.dtypes
```

```python
cleaned_df.info()
```

```python
#Coverting the data type
cleaned_df['budget'] = cleaned_df['budget'].fillna(0).astype(int)  # Replace NaN with 0 and then convert to int
cleaned_df['gross'] = cleaned_df['gross'].fillna(0).astype(int) # Replace NaN with 0 and then convert to int
```

```python
#Adding a new column
cleaned_df['yearcorrect'] = cleaned_df['released'].str.extract(r'(\d{4})')
```

```python
cleaned_df
```

```python
cleaned_df.info()
```

```python
cleaned_df['yearcorrect'] = cleaned_df['yearcorrect'].astype(int)
```

```python
#Sorting by highest grossing movies
cleaned_df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
#Converting the data type for Votes
cleaned_df['votes'] = cleaned_df['votes'].astype(int)
```

```python
#Scatter plot 
plt.scatter(x=cleaned_df['budget'], y=cleaned_df['gross'])
plt.title("Budget vs Gross Earnings")
plt.xlabel("Gross Earnings")
plt.ylabel("Movie Budget")
plt.show()
```
